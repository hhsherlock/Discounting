```python
x = np.linspace(0,12.2, 1000)
# y = a_higher/(1+b_higher/x)

a_cafe = np.exp(0.5)
b_cafe = np.exp(3.25)
# a_gamble = np.exp(0.625)
a_gamble = np.exp(1)
b_gamble = np.exp(3.5)

y_cafe = 1/(1+b_cafe*np.exp(-a_cafe*x))
y_gamble = 1/(1+b_cafe*2**(-a_cafe*x))

# y = c_higher/(1+b_higher*np.exp(-a_higher*x))

plt.xlabel('delay')
plt.ylabel('time perception')
plt.plot(x,y_cafe, label = 'cafe_big', color='royalblue')
plt.plot(x,y_gamble, label = 'gamble_big', color='orange')

# plt.plot(x,y_cafe, label='cafe')
# plt.plot(x,y_gamble, label='gamble')
plt.legend(loc='upper right', bbox_to_anchor=(1.4, 1))
```
