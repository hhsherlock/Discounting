```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Mon Feb 26 2024

@author: Yaning
"""

from matplotlib import pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
mean_u_cafe = 0.68252325
# mean_u_cafe_v = 0.0385
sigma_u_cafe = 3.9410226
sigma_es_cafe = 2.824972

mean_u_gamb = 1.0304108
sigma_u_gamb = 3.069599
sigma_es_gamb = 3.6164086

ll_value = 50
delays = np.linspace(1, 122, 1000)
x = np.linspace(-20, 80, 1000)
delay = 50

estimation_cafe = []
for i in delays:
    estimation_value = (mean_u_cafe*0.1*i*sigma_es_cafe**2 + ll_value*sigma_u_cafe**2)/(
            0.1*i*sigma_es_cafe**2 +sigma_u_cafe**2)
    estimation_cafe.append(estimation_value)

estimation_gamb = []
for i in delays:
    estimation_value = (mean_u_gamb*0.1*i*sigma_es_gamb**2 + ll_value*sigma_u_gamb**2)/(
            0.1*i*sigma_es_gamb**2 +sigma_u_gamb**2)
    estimation_gamb.append(estimation_value)

# # those are the final distributions
# distr_cafe = stats.norm.pdf(x, (mean_u_cafe*0.1*delay*sigma_es_cafe**2 + ll_value*sigma_u_cafe**2)/(
#             0.1*delay*sigma_es_cafe**2 +sigma_u_cafe**2), ((sigma_u_cafe**2*0.1*delay*sigma_es_cafe**2)/(
#                 sigma_u_cafe**2 + 0.1*delay*sigma_es_cafe**2))**0.5)

# distri_gamb = stats.norm.pdf(x, (mean_u_gamb*0.1*delay*sigma_es_gamb**2 + ll_value*sigma_u_gamb**2)/(
#             0.1*delay*sigma_es_gamb**2 +sigma_u_gamb**2), ((sigma_u_gamb**2*0.1*delay*sigma_es_gamb**2)/(
#                 sigma_u_gamb**2 + 0.1*delay*sigma_es_gamb**2))**0.5)
```

```python
# set the plot structure
fig = plt.figure(constrained_layout=False, figsize=(5,5))
gs = fig.add_gridspec(nrows=1, ncols=1, left=0.05, right=0.95, wspace=0.35, hspace=0.35)

ax1 = fig.add_subplot(gs[0], xlabel = 'delay', ylabel = 'estimation')
# ax2 = fig.add_subplot(gs[1], xlabel = 'value', ylabel = 'pdf')


ax1.plot(delays, estimation_cafe, label = 'cafe')
ax1.plot(delays, estimation_gamb, label = 'gamble')
ax1.legend(loc="upper right")





plt.show()
```
