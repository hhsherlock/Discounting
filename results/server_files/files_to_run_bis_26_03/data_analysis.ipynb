```python
file = open('Documents/pyro_models/files_to_run/intertemporal_choice_dataset_all_trials.csv')
```

```python
csvreader = csv.reader(file)
```

```python
header = []
header = next(csvreader)
# print(header)
```

```python
# [[[[p],[p],...version1], [[p],[p],...version2]context1], [[[p],[p],...version1], [[p],[p],...version2]context2]]
# the amounts of participants in s1 c1, s2 c2 are different from s1 c2, s2 c1. 
# to keep all amounts the same for array conversion
# so he two arrays are for the addtional four participants.
data = [[[],[]], [[],[]]]
n_trial = 170
add_c1_s1 = []
add_c2_s2 = []

for row in csvreader:
    if row[11] == '0' and row[9] == 'S1':
        individual = []
        individual.append(row[1:7] + row[10:])
        for i in range(n_trial-1):
            next_row = next(csvreader)
            kurz_next_row = next_row[1:7] + next_row[10:]
            individual.append(kurz_next_row)
        if len(data[0][0]) >= 14:
            add_c1_s1.append(individual)
        else:
            data[0][0].append(individual)
    elif row[11] == '1' and row[9] == 'S1':
        individual = []
        individual.append(row[1:7] + row[10:])
        for i in range(n_trial-1):
            next_row = next(csvreader)
            kurz_next_row = next_row[1:7] + next_row[10:]
            individual.append(kurz_next_row)
        data[1][0].append(individual)
    elif row[11] == '0' and row[9] == 'S2':
        individual = []
        individual.append(row[1:7] + row[10:])
        for i in range(n_trial-1):
            next_row = next(csvreader)
            kurz_next_row = next_row[1:7] + next_row[10:]
            individual.append(kurz_next_row)
        data[0][1].append(individual)
    elif row[11] == '1' and row[9] == 'S2':
        individual = []
        individual.append(row[1:7] + row[10:])
        for i in range(n_trial-1):
            next_row = next(csvreader)
            kurz_next_row = next_row[1:7] + next_row[10:]
            individual.append(kurz_next_row)
        if len(data[1][1]) >= 14:
            add_c2_s2.append(individual)
        else:
            data[1][1].append(individual)
```

```python
for i in range(len(data)):
    for j in range(len(data[0])):
        for k in range(len(data[0][0])):
            for l in range(len(data[0][0][0])):
                for m in range(len(data[0][0][0][0])):
                    data[i][j][k][l][m] = float(data[i][j][k][l][m])
```
